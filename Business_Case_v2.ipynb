```python
# Merging the datasets
# First, we need to mark the invoice data with a flag indicating whether it falls within a promotional period

# Create a copy of the promo data with only necessary columns for merging
promo_relevant = promo_data[['basecode', 'location_cd', 'ship_start_dt', 'ship_end_dt']].drop_duplicates()

# Merge the invoice data with the promo data to flag promotional sales
# We will use a left join to keep all invoice records and only mark those that match promotional criteria
merged_data = pd.merge(invoice_data_cleaned, promo_relevant, how='left', left_on=['basecode', 'LOCATION_CD'], right_on=['basecode', 'location_cd'])

# Create a new column to flag promotional sales
merged_data['is_promo'] = ((merged_data['INVOICE_DT'] >= merged_data['ship_start_dt']) & (merged_data['INVOICE_DT'] <= merged_data['ship_end_dt']))

# Drop duplicate and unnecessary columns from the merge
merged_data.drop(columns=['location_cd', 'ship_start_dt', 'ship_end_dt'], inplace=True)

# Checking the result of the merge
merged_data.head(),
```

```python
# Calculate the average sales volume during promotional and non-promotional periods
average_sales_by_promo_status = merged_data.groupby('is_promo')['invoice_qty'].mean().reset_index()

# Calculate the total sales volume by brand and promo status
total_sales_by_brand_and_promo = merged_data.groupby(['brand_cd', 'is_promo'])['invoice_qty'].sum().reset_index()

# Calculate the total sales volume by customer segment and promo status
total_sales_by_customer_and_promo = merged_data.groupby(['customer_hierarchy_lvl2_cd', 'is_promo'])['invoice_qty'].sum().reset_index()

average_sales_by_promo_status, total_sales_by_brand_and_promo.head(), total_sales_by_customer_and_promo.head()
```

```python
promo_only = merged_data[merged_data['is_promo'] == True]
promo_only.reset_index(inplace=True, drop=True)
promo_only.head()
```

```python
## transforming in sigle value by each bill

agg_funcs = {'invoice_qty': 'sum'}  # Summing invoice quantities
# For columns that should not be summed but kept, we use 'first' to keep the first occurrence
for col in promo_only.columns.difference(['INVOICE_CD', 'invoice_qty']):
    agg_funcs[col] = 'first'

promo_sales_sum_by_invoice = promo_only.groupby('INVOICE_CD', as_index=False).agg(agg_funcs)

promo_sales_sum_by_invoice


```

```python
# Sum invoice quantities by quarter
quarterly_data = promo_sales_sum_by_invoice.resample('Q').sum()

# You might also want to keep other relevant aggregated information, adjusting the aggregation method as needed
# For example, using 'first' for categorical data that doesn't need summing but should be retained per quarter
agg_funcs = {'invoice_qty': 'sum', 'MATERIAL_NUM_CD': 'first', 'SALESORG_CD': 'first', 'is_promo': 'max'}
quarterly_data = promo_sales_sum_by_invoice.resample('Q').agg(agg_funcs)

# Display the quarterly aggregated data
print(quarterly_data)
```

```python
data = quarterly_data['invoice_qty']
data
```

```python
# ==============================================================================
steps = 2
data_train = data[:-steps]
data_test  = data[-steps:]
```

```python
fig, ax = plt.subplots(figsize=(6, 3))
data_train.plot(ax=ax, label='train')
data_test.plot(ax=ax, label='test')
ax.legend();
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from skforecast.ForecasterAutoreg import ForecasterAutoreg
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Create and fit forecaster
# ==============================================================================
forecaster = ForecasterAutoreg(
                 regressor = RandomForestRegressor(random_state=123),
                 lags      = 2
             )

forecaster.fit(y=data_train)
forecaster
```

```python
# Predict
# ==============================================================================
predictions = forecaster.predict(steps=2)
predictions.head()
```

```python
# Plot predictions
# ==============================================================================
fig, ax = plt.subplots(figsize=(6, 3))
data_train.plot(ax=ax, label='train')
data_test.plot(ax=ax, label='test')
predictions.plot(ax=ax, label='predictions')
ax.legend();
```

```python
# Prediction error
# ==============================================================================
error_mse = mean_squared_error(
                y_true = data_test,
                y_pred = predictions
            )
print(f"Test error (mse): {error_mse}")
```

```python

# Predict intervals
# ==============================================================================
predictions = forecaster.predict_interval(steps=4, interval=[5, 95], n_boot=100)
predictions.head(3)
```

```python
# Plot prediction interval
# ==============================================================================
fig, ax = plt.subplots(figsize=(6, 3))
data_test.plot(ax=ax, label="test")
predictions['pred'].plot(ax=ax, label="prediction")
ax.fill_between(
    predictions.index,
    predictions['lower_bound'],
    predictions['upper_bound'],
    color = 'blue',
    alpha = 0.3,
    label = '80% interval'
)
ax.legend(loc="upper left");
```
